```python
def impute_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    if pd.isnull(Age):
        if Pclass==1:
            return 37
        elif Pclass==2:
            return 29
        else:
            return 24
    else:
        return Age         
```

```python
# axis=0 would cause the fxn to be applied to the row
#axis =1 refers to the column
df1['Age']= df1[['Age','Pclass']].apply(impute_age,axis=1)
```
